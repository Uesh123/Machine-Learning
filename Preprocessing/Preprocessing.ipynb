# Preprocessing
- missing data handling
    1. imputation
        - numarical(mean)
        - categorical (text)-mode(most frequent)
    2.removal
        - only when too much of data is missing from the columns
        - remove row when it is completely empty
        - remove rows when dependent value if missing
- conversion (encoding)
    1. Lable encoding
    2. OneHotEncoding
- normalization(scaling)
    1. minmax
    2. standard


```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import missingno as mso
```

```python
data= pd.DataFrame(
    {
        'name':['Ajay','Vijay','sujay','sanjay','alex'],
        'age':[20,30,np.nan,np.nan,29],
        'iq':[100,np.nan,90,50,np.nan],
        'married': [True,True,False,np.nan,np.nan]
    }
)
```

```python
data
```

If the columns had a independing uniqe value is not usefull

```python
mso.matrix(data)
```

1. import class 
2. create instance (object)
3. pass data (fit)
4. get output (transform)
5. combine (fit transform())

```python
si=SimpleImputer() #object
data[['age','iq']]
```

```python
si.fit_transform(data[['age','iq']])
```

```python

data[['age','iq']]= si.fit_transform(data[['age','iq']])
data
```

```python
si2=SimpleImputer(strategy='most_frequent')
si2.fit_transform(data[['married']])
```

```python
# to save the data
data[['married']]=si2.fit_transform(data[['married']])
```
